## Function to get tags from a single text.

```python
def get_tags(text):
    tags = []
    
    pattern_1 = "[a-zA-Z]{2}[a-zA-Z]{0,2}[-][0-9]{8}[a-zA-Z]{0,1}"
    pattern_2 = "[a-zA-Z]{2}[a-zA-Z]{0,2}[-][0-9]{8}[a-zA-Z]+[/a-zA-Z]+" 
    pattern_3 = "[a-zA-Z]{2}[a-zA-Z]{0,2}[-][0-9]{0,8}[/0-9]+"
    
    matches_1 = re.findall(pattern_1, text)
    matches_2 = re.findall(pattern_2, text)
    matches_3 = re.findall(pattern_3, text)
    
    if len(matches_3)> 0:
        for match in matches_3:
            sep = match.split('/')
            if len(sep) > 1:
                sep_1 = sep[0]
                tags.append(sep_1)
                for i in range(1, len(sep)):
                    l = -len(sep[i])
                    temp = sep_1[:l] + sep[i]
                    tags.append(temp)
                    
    if len(matches_2)> 0:
#         print(matches_2)
        for match in matches_2:
            sep = match.split('/')
            if len(sep) > 1:
                sep_1 = sep[0]
                tags.append(sep_1)
                for i in range(1, len(sep)):
                    l = -len(sep[i])
                    temp = sep_1[:l] + sep[i]
                    tags.append(temp)
                    
    if len(matches_1)> 0:
        tags = tags+matches_1
        
    tags_text = ";".join(list(dict.fromkeys(tags)))
    return tags_text
```

## Getting the First tag in the list of tags in CTDB Tags column in Connection table

```python
conn['CTDB TAG'] = conn['CTDB TAG'].apply(lambda x: x.split(';')[0])
conn
```

```python
# Many Reports in Report Tag has no value from CTDB table in these cases we have left the Functional and Physical class columns as 'NaN' to perserve the original number of rows in the report table.
# Note: This part of the code could have been done in one line. But it was easier to document this way.

result = rep.drop(['Functional Class', 'Physical Class'], axis=1)                                          # Step-1: Remove the Functional and Physical class columns
result = result.merge(conn, how = 'left', left_on = 'Report Tag', right_on = 'REPORT Tag')                 # Step-2: Left Join report and connection tables on the 'Report Tag' columns
result = result.merge(ctdb, how = 'left', left_on = 'CTDB TAG', right_on = 'CTDB Tag')                     # Step-3: Left Join previously combined tables with the CTDB table on the 'CTDB Tag' columns
result = result.drop(['REPORT Tag','Rule', 'CTDB TAG', 'CTDB Tag', 'Tag Description', 'Tag Type'], axis=1) # Step-4: Only keep the columns that were in the original Report Table and remove other columns.
result
```

```python
result.to_excel("Task 2 output.xlsx", index=False)
```
